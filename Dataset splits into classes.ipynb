```python
import os
import shutil

path = 'D:\\ConcordiaU\\Winter 23\\AI\\Project\\Datasets\\1. APPA-REAL\\Cropped Faces copy from kaggle\\final_files\\'

os.makedirs(os.path.join(path, '0'), exist_ok=True)
os.makedirs(os.path.join(path, '1'), exist_ok=True)
os.makedirs(os.path.join(path, '2'), exist_ok=True)
os.makedirs(os.path.join(path, '3'), exist_ok=True)
os.makedirs(os.path.join(path, '4'), exist_ok=True)

for index, row in df_classes.iterrows():
    file_path = row['file_path']
    classes = row['Classes']

   
    if classes == 0:
        shutil.move(file_path, os.path.join(path, '0', os.path.basename(file_path)))
    elif classes == 1:
        shutil.move(file_path, os.path.join(path, '1', os.path.basename(file_path)))
    elif classes == 2:
        shutil.move(file_path, os.path.join(path, '2', os.path.basename(file_path)))
    elif classes == 3:
        shutil.move(file_path, os.path.join(path, '3', os.path.basename(file_path)))
    elif classes == 4:
        shutil.move(file_path, os.path.join(path, '4', os.path.basename(file_path)))
```
